```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
from matplotlib import figure
sns.set_theme()
```

```python
#load and read csv

df=pd.read_csv("world-happiness-report-2021.csv")
df.head()
```

```python
#rename the columns and read first 5 rows

df.rename(columns={'Country name':'country','Regional indicator':'region','Ladder score':'score', \
                  'Standard error of ladder score':'stderr_score','Logged GDP per capita':'gdp', \
                   'Social support':'social_support','Healthy life expectancy':'life_expectancy',\
                  'Freedom to make life choices':'freedom','Generosity':'generosity',\
                  'Perceptions of corruption':'corruption_perception',},inplace=True)
```

```python
df.head()
```

```python
#drop columns and display first 5 rows

df.columns
```

```python
df.drop(['Ladder score in Dystopia',\
       'Explained by: Log GDP per capita', 'Explained by: Social support',\
       'Explained by: Healthy life expectancy',\
       'Explained by: Freedom to make life choices',\
       'Explained by: Generosity', 'Explained by: Perceptions of corruption',\
       'Dystopia + residual'],axis=1, inplace=True)
df.head()
```

```python
# review correlation between features

df.corr()
```

```python
#create dataframe by regions, and score against social support

df1=df.loc[df['region']=='Western Europe',['region','score','social_support']]
df2=df.loc[df['region']=='North America and ANZ',['region','score','social_support']]
df3=df.loc[df['region']=='Middle East and North Africa',['region','score','social_support']]
df4=df.loc[df['region']=='Latin America and Caribbean',['region','score','social_support']]
df5=df.loc[df['region']=='Central and Eastern Europe',['region','score','social_support']]
df6=df.loc[df['region']=='East Asia',['region','score','social_support']]
df7=df.loc[df['region']=='Southeast Asia',['region','score','social_support']]
df8=df.loc[df['region']=='Commonwealth of Independent States',['region','score','social_support']]
df9=df.loc[df['region']=='Sub-Saharan Africa',['region','score','social_support']]
df10=df.loc[df['region']=='South Asia',['region','score','social_support']]
```

```python
#plot average happiness according regions

df1=df.groupby('region')['score'].agg(np.mean).to_frame()
df1=df1.reset_index()
df1.sort_values(by=['score'], axis=0, ascending=False,inplace=True)
sns.barplot(x='score', y='region', data=df1);
plt.xlabel('happiness score');
```

```python
#plot average social_support (family) according regions

df1=df.groupby('region')['social_support'].agg(np.mean).to_frame()
df1=df1.reset_index()
df1.sort_values(by=['social_support'], axis=0, ascending=False,inplace=True)
sns.barplot(x='social_support', y='region', data=df1);
plt.xlabel('social support (family)');
```

```python
#plot average healthy life expectancy according regions

df1=df.groupby('region')['life_expectancy'].agg(np.mean).to_frame()
df1=df1.reset_index()
df1.sort_values(by=['life_expectancy'], axis=0, ascending=False,inplace=True)
sns.barplot(x='life_expectancy', y='region', data=df1);
plt.xlabel('healthy life expectancy');
```

```python
#plot average freedom to make choices according regions

df1=df.groupby('region')['freedom'].agg(np.mean).to_frame()
df1=df1.reset_index()
df1.sort_values(by=['freedom'], axis=0, ascending=False,inplace=True)
sns.barplot(x='freedom', y='region', data=df1);
plt.xlabel('healthy life expectancy');
```

```python
#plot average generosity according regions

df1=df.groupby('region')['generosity'].agg(np.mean).to_frame()
df1=df1.reset_index()
df1.sort_values(by=['generosity'], axis=0, ascending=False,inplace=True)
sns.barplot(x='generosity', y='region', data=df1);
plt.xlabel('generosity');
```

```python
#create dataframe for South East Asia region, country and happiness score

df2=df.loc[df['region']=='Southeast Asia',['country','score']]
df2.sort_values(by=['score'],axis=0, ascending=False, inplace=True)

sns.barplot(x='score', y='country', data=df2);
plt.xlabel('happiness score');
plt.ylabel('South East Asia');
```
