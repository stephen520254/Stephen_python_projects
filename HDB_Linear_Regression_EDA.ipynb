```python
# import python libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
sns.set_theme()
plt.rcParams['figure.figsize']=[10,8]
import datetime
```

```python
#load and read csv files

yr1999 = pd.read_csv('resale-flat-prices-based-on-approval-date-1990-1999.csv')
yr2012=pd.read_csv('resale-flat-prices-based-on-approval-date-2000-feb-2012.csv')
yr2016=pd.read_csv('resale-flat-prices-based-on-registration-date-from-jan-2015-to-dec-2016.csv')
yr2021=pd.read_csv('resale-flat-prices-based-on-registration-date-from-jan-2017-onwards.csv')
yr2014=pd.read_csv('resale-flat-prices-based-on-registration-date-from-mar-2012-to-dec-2014.csv')
yr1999.head(2)
```

```python
# drop 'remaining_lease' column from yr2016 & yr2021 dataframes

col =['remaining_lease']
file_list = [yr2016,yr2021]
for w in file_list:
    for i in col:
        w.drop(i,axis='columns',inplace=True)
```

```python
# drop 'block', 'street_name', 'flat_model' and 'storey_range' columns from all dataframes (df : yr2021)

col =['block','street_name','flat_model','storey_range']
file_list = [yr1999,yr2012,yr2014,yr2016,yr2021]
for w in file_list:
    for i in col:
        w.drop(i,axis='columns',inplace=True)
```

```python
#concatenate into 5 dataframes (yr1999 : yr2021) into 1 dataframe yr_combine

file_list = [yr1999,yr2012,yr2014,yr2016,yr2021]
yr_combine = pd.concat(file_list)
```

```python
# review yr_combine dataframe 1st 5 rows

yr_combine.head()
```

```python
#convert 'lease_commence_date' to datetime format

from datetime import datetime

date_format=[]
date_list = yr_combine['lease_commence_date'].astype(str).values.tolist()
for j in date_list:
    date_obj = datetime.strptime(j, '%Y')
    date_format.append(date_obj)
yr_combine['lease_commence_date']=date_format
```

```python
#convert month to datetime format and assign to new column 'month_dt'

from datetime import datetime

date_format=[]
date_list = yr_combine['month'].astype(str).values.tolist()
for i in date_list:
    date_obj = datetime.strptime(i, '%Y-%m')
    date_format.append(date_obj)
yr_combine['month_dt']=date_format
```

```python
# compute days expired by subtracting 'month_dt' by 'lease_commence_date'.  This will give you the age of the flat.

yr_combine['days_expired'] = yr_combine['month_dt']-yr_combine['lease_commence_date']
```

```python
# compute remaining_lease by converting 'days_expired' into integer in years, subtracting it from 99 years.

yr_combine['remaining_lease'] = (99-((pd.to_numeric(yr_combine['days_expired'].dt.days, downcast='integer'))/365)).round(decimals=0)
```

```python
# convert 'month' format to integer, removing the month part of the string to become 'yyyy'

yr_combine['month']= [int(i[:-3]) for i in yr_combine['month']]
```

```python
#replace flat_type value of 'MULTI-GENERATION' and 'MULTI GENERATION' to 'MULTI_GENERATION'. 

yr_combine['flat_type'].replace('MULTI-GENERATION','MULTI_GENERATION',inplace=True)
yr_combine['flat_type'].replace('MULTI GENERATION','MULTI_GENERATION',inplace=True)
```

```python
#create a new column "price_per_sqm"
#rename column name 'month' to 'year'


yr_combine['price_per_sqm']=round(yr_combine['resale_price']/yr_combine['floor_area_sqm'],0)
yr_combine.rename(columns={'month':'year'}, inplace=True)
yr_combine.head()
```

```python
yr_grt2020 = yr_combine[yr_combine['year']>2020]
yr_grt2020.head()
```

### May I know the trend of  HDB this year?  Will the price drop? Should I wait?

```python
from sklearn.preprocessing import StandardScaler
```

```python
#find average HDB price by year and standardize the data

avg_price = yr_combine.groupby('year')['price_per_sqm'].agg(np.mean).to_frame().reset_index()
scaler = StandardScaler()
x=avg_price['price_per_sqm'].values.reshape(-1,1)
avg_price['price_per_sqm_sc']=scaler.fit_transform(x)
```

```python
#find the total amount of resale by year and standardize the data

total_resale_amt = yr_combine.groupby('year')['resale_price'].agg(np.sum).reset_index()
scaler = StandardScaler()
x=total_resale_amt['resale_price'].values.reshape(-1,1)
total_resale_amt['resale_price_sc']=scaler.fit_transform(x)
```

```python
# plot a regression line for price per sqm
from scipy import stats

x=np.array(avg_price['year'])
y=np.array(avg_price['price_per_sqm_sc'])

slope, intercept, r, p, stderr = stats.linregress(x,y)

func = lambda x: slope*x + intercept

x1=np.array([1990,2021])
y1=np.array([func(1990), func(2021)])
```

```python
# plot a regression line for total resale volume
from scipy import stats

x2=np.array(total_resale_amt['year'])
y2=np.array(total_resale_amt['resale_price_sc'])

slope, intercept, r, p, stderr = stats.linregress(x2,y2)

func = lambda x: slope*x + intercept

x3=np.array([1990,2021])
y3=np.array([func(1990), func(2021)])
```

```python
#plot price per sqm and total resale amount and regression line for price per sqm

plt.plot(avg_price['year'],avg_price['price_per_sqm_sc'], linestyle='--',linewidth=4);
plt.plot(x1,y1, linestyle='--',linewidth=4)
plt.plot(total_resale_amt['year'], total_resale_amt['resale_price_sc']);

plt.plot(x3,y3)

fig=plt.gcf();
fig.set_size_inches(15,11);
plt.xlabel('Year');
plt.ylabel('Trend');
plt.title('Total Resale Amount VS Price Per Sqm Based on Year', fontsize=15);
plt.legend(['Price Per Sqm','Prediction For Price Per Sqm','Total Resale Amount','Prediction For Total Resale Amount'],\
           fontsize=15, loc='lower right');
```

### Can we compare the trend of private properties vs HDB in resale prices?

```python
private2020 = pd.read_csv('private-residential-property-price-index-base-quarter-2009-q1-100.csv')
```

```python
# format year column

private2020['year']=[i[:-3] for i in private2020['quarter']]
```

```python
# re-ordering column names

private2020[['year','quarter','level_1','value']]
```

```python
# compute average price by year

private_avg_price = private2020.groupby('year')['value'].agg(np.mean).to_frame()
private_avg_price.reset_index(inplace=True)
```

```python
# format year to integter

private_avg_price['year']=[int(j) for j in private_avg_price['year']]
```

```python
# create dataframe for transactions after 1989

private_avg_price = private_avg_price[private_avg_price['year']>1989]

private_avg_price.head()
```

```python
# standardize the chart

scaler = StandardScaler()
x=private_avg_price.value.values.reshape(-1,1)
private_avg_price['value1']=scaler.fit_transform(x)
```

```python
# remove year 2021 from dataframe

avg_price_before_2021=avg_price[avg_price['year']<2021]
```

```python
x5=np.array(avg_price_before_2021['year'])
y5=np.array(avg_price_before_2021['price_per_sqm_sc'])

slope, intercept, r, p, stderr = stats.linregress(x5,y5)
func = lambda x: slope*x + intercept

x6=np.array([1990,2020])
y6=np.array([func(1990), func(2020)])
```

```python
x7=np.array(private_avg_price['year'])
y7=np.array(private_avg_price['value1'])

slope1, intercept1, r, p, stderr = stats.linregress(x7,y7)
func1 = lambda x: slope1*x + intercept1

x8=np.array([1990,2020])
y8=np.array([func1(1990), func1(2020)])
```

```python
# create dataframe for HDB for year before 2021 to align with private property price index

fig=plt.gcf()
fig.set_size_inches(15,11)

plt.plot(avg_price_before_2021['year'],avg_price_before_2021['price_per_sqm_sc'],c = 'g', linestyle='--',linewidth=4);
plt.plot(x6,y6,linestyle ='--',linewidth=4 );
plt.plot(private_avg_price['year'],private_avg_price['value1']);
plt.plot(x8,y8,);

plt.xlabel('Year',fontsize=15);
plt.ylabel('Price',fontsize=15);
plt.title('HDB VS Private Property By Year',fontsize=15);
plt.legend(['HDB','Prediction HDB','Private Property','Prediction Private Property'], fontsize=15);
```
