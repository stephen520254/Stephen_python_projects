```python
from sklearn.cluster import KMeans
```

Classification of total rainfall from 1982 to present

```python
df5['month'] = [int(i[:-3]) for i in df5['month']]
```

```python
k = range(1,10)
sse=[]
for i in k:
    km = KMeans(n_clusters=i, random_state=1).fit(df5)
    sse.append(km.inertia_)
```

```python
fig = plt.gcf();
fig.set_size_inches(10,8);
plt.plot(k,sse);
plt.xlabel('Number Of Clusters (k)');
plt.ylabel('Sum of Squared Error');
plt.title('Elbow Techinque To Compute K');
```

```python
df5['clusters'] = KMeans(n_clusters=3, random_state=1).fit_predict(df5)
```

Classification of maximum temperature from 1982 to now

```python
df6.month = [int(i[:-3]) for i in df6.month]
```

```python
# run elbow technique to find optimum number of clusters

sse=[]
k = range(1,10)
for i in k:
    km = KMeans(n_clusters=i, random_state=2).fit(df6)
    sse.append(km.inertia_)

fig = plt.gcf();
fig.set_size_inches(12,8);
plt.plot(k,sse);
```
